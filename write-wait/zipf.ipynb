Explore the Zipfian distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import zeta  
```

```python
for a in np.linspace(1.5, 9., 6):
    n=1.
    x_max = 25
    k = np.arange(1, x_max + 1)
    plt.plot(k, n*(k**-float(a))/zeta(a), '.-', alpha=0.5,
             label=f'a={a}')   
plt.grid(alpha=0.4)
plt.legend()
plt.title(f'Zipf histogram')
plt.show()
```
